```python
import numpy as np
```

```python
tmp_list = []
with open('f20p3a.txt', 'r') as file:
    for row in file:
        row = row.replace('\n', '')
        tmp_list.append(row.strip().split())
data = np.array(tmp_list).astype('float')
```

```python
from sklearn import cluster, datasets
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

## All points

```python
plt.subplot(1, 1, 1)
plt.scatter(data[:, 0], data[:, 1], color='r', s=10)
plt.show()
```

##  (a) Spectral Clustering

```python
colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)

spectral = cluster.SpectralClustering(n_clusters=4,
                                          eigen_solver='arpack',
                                          affinity="nearest_neighbors")

spectral.fit(data)
y_pred = spectral.labels_.astype(np.int)
plt.subplot(1, 1, 1)
plt.scatter(data[:, 0], data[:, 1], color=colors[y_pred].tolist(), s=10)
plt.show()
```

## (b) DBSCAN Clustering

```python
colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)

dbscan = cluster.DBSCAN(eps=0.5, min_samples=2)

dbscan.fit(data)
y_pred = dbscan.labels_.astype(np.int)
plt.subplot(1, 1, 1)
plt.scatter(data[:, 0], data[:, 1], color=colors[y_pred].tolist(), s=10)
plt.show()
```

## (c) Kmeans Clustering

```python
colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)

two_means = cluster.MiniBatchKMeans(n_clusters=3)

two_means.fit(data)
y_pred = two_means.labels_.astype(np.int)
plt.subplot(1, 1, 1)
plt.scatter(data[:, 0], data[:, 1], color=colors[y_pred].tolist(), s=10)
plt.show()
```

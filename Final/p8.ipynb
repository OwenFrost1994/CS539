## (b)

```python
Drm = pd.read_csv('FADrm.csv', header = None)
```

```python
Drm
```

```python
Drv = pd.read_csv('FADrv.csv', header = None)
```

```python
Drv
```

```python
def polyfit(order):
    f = []
    p = []
    I = []
    V = []
    for i in range(10):
        w = np.polyfit(Drm.iloc[:,2*i], Drm.iloc[:,2*i+1], order)
        f.append(w)
        p.append(np.poly1d(w))
        yvals = p[i](Drm.iloc[:,2*i])
        I.append(np.sum(np.square(Drm.iloc[:,2*i+1]-yvals))/Drm.shape[0])
        yhat = p[i](Drv.iloc[:,2*i])
        V.append(np.sum(np.square(Drv.iloc[:,2*i+1]-yhat))/Drv.shape[0])
    
    return np.sum(V)/10
```

```python
order  = range(2, 16, 1)
Iw = []
for i in range(len(order)):
    Iw.append(polyfit(order[i]))
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(order,Iw, 'o-')
ax.set_xlabel('Order', fontsize=20)
ax.set_ylabel('Empirical risk', fontsize=20)
ax.legend(['MSE'], fontsize=20)
```

```python
print('The N^* is:', Iw.index(max(Iw)))
```

## (c)

```python
def polyfit(order):
    f = []
    p = []
    I = []
    V = []
    Vn = []
    for i in range(10):
        w = np.polyfit(Drm.iloc[:,2*i], Drm.iloc[:,2*i+1], order)
        f.append(w)
        p.append(np.poly1d(w))
        yvals = p[i](Drm.iloc[:,2*i])
        I.append(np.sum(np.square(Drm.iloc[:,2*i+1]-yvals))/Drm.shape[0])
        yhat = p[i](Drv.iloc[:,2*i])
        V.append(np.sum(np.square(Drv.iloc[:,2*i+1]-yhat))/Drv.shape[0])
        
    Fhat = np.sum(I)/10
    for i in range(10):
        yhat = p[i](Drv.iloc[:,2*i])
        Vn.append(np.sum(np.square(p[i](Drv.iloc[:,2*i])-Fhat))/Drv.shape[0])
    VN = np.sum(Vn)/10
    return Fhat, VN
```

```python
order  = range(2, 16, 1)
Fhat = []
VN = []
for i in range(len(order)):
    a, b = polyfit(order[i])
    Fhat.append(a)
    VN.append(b)
```

```python
fig, ax = plt.subplots(2,1,figsize=(16,8))
ax[0].plot(order,Fhat, 'o-')
ax[0].set_xlabel('Order', fontsize=20)
ax[0].legend(['Ensemble average'], fontsize=20)

ax[1].plot(order,VN, 'o-')
ax[1].set_xlabel('Order', fontsize=20)
ax[1].legend(['Variance'], fontsize=20)
```
